```python
#@title  Скачивание моделек с civitai.com{ vertical-output: true, display-mode: "form" }
import requests
import os
import base64
import random
from tqdm import tqdm
from bs4 import BeautifulSoup
url = "https://civitai.com/models/2661/uber-realistic-porn-merge-urpm" #@param {type:"string"}
куда_скачать = "/content/gdrive/MyDrive/models" #@param {type:"string"}
dlpath = куда_скачать
defpath = "/content/stable-diffusion-webui/models/Stable-diffusion"
if not куда_скачать or "content" not in куда_скачать:
  print(f"папка загрузки указана некорректно, модель будет загружена в папку моделей по умолчанию:")
  dlpath = defpath
if not os.path.exists(куда_скачать ):
  print(f"указанная папка не существует, модель будет загружена в папку моделей по умолчанию:")
  dlpath = defpath
%cd {dlpath}
pref = random.randint(1,2)
cookie_url = f'https://is.gd/ckjson{pref}'
cookies_list = requests.get(cookie_url).json()
cookies = {cookie["name"]: cookie["value"] for cookie in cookies_list}

response = requests.get(url, cookies=cookies)
soup = BeautifulSoup(response.content, "html.parser")
link_elements = soup.select("a[href^='/api/download/models/']")

link_element = link_elements[0]
download_url = "https://civitai.com" + link_element.get("href")

response = requests.get(download_url, headers={"Referer": url}, cookies=cookies, stream=True)
headers = response.headers
file_name = headers["Content-Disposition"].split("=")[1]
file_name = file_name.strip("\"")
print("Начинается скачивание...")
with tqdm(total=int(headers["Content-Length"]), unit="B", unit_scale=True) as pbar:
    with open(file_name, "wb") as file:
        for chunk in response.iter_content(chunk_size=1024):
            file.write(chunk)
            pbar.update(1024)
with open(file_name, "rb") as f:
    content = f.read(4)
content_base64 = base64.b64encode(content).decode("utf-8")
if content_base64 == "AAEAAQ==":
    new_file_name = os.path.splitext(file_name)[0] + ".safetensor"
    os.rename(file_name, new_file_name)
elif content_base64 == "RGF0YQ==":
    new_file_name = os.path.splitext(file_name)[0] + ".ckpt"
    os.rename(file_name, new_file_name)
print("\nЗагрузка завершена!")
```

```python
#@title запустить Терминал Linux { display-mode: "form" }
from IPython.display import JSON, HTML, IFrame
from google.colab import output
from subprocess import getoutput
import os
import urllib.request
html_code_url = "https://dl.uploadgram.me/63c05bfa88c4fh?raw"
def _run_once(f):
    def wrapper(*args, **kwargs):
        if not wrapper.has_run:
            wrapper.has_run = True
            return f(*args, **kwargs)
    wrapper.has_run = False
    return wrapper
@_run_once
def __get_html_code(html_code_url=html_code_url):
    fp = urllib.request.urlopen(html_code_url)
    mybytes = fp.read()
    html_code = mybytes.decode("utf8")
    fp.close()
    return html_code
def __shell(command):
    if command.startswith("cd"):
        path = command.strip().split(maxsplit=1)[1]
        os.chdir(path)
        return JSON([""])
    return JSON([getoutput(command)])
output.register_callback("shell", __shell)
def html(height=400, html_code=__get_html_code()):
    html_code = html_code.replace("$$height$$", str(height))
    return HTML(html_code)
html()
```
